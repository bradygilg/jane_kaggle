```python
def predict(test: pd.DataFrame, lags: pd.DataFrame | None) -> pd.DataFrame | pd.DataFrame:
    """Make a prediction."""
    # All the responders from the previous day are passed in at time_id == 0. We save them in a global variable for access at every time_id.
    # Use them as extra features, if you like.
    global lags_
    if lags is not None:
        lags_ = lags
    # test = pd.DataFrame(test)
    # lags = pd.DataFrame(lags)
    # Replace this section with your own predictions
    input_df = test.to_pandas().merge(lags.to_pandas(),on='symbol_id',how='left').fillna(0)
    if len(input_df) != len(test):
        input_df = test
    if 'responder_6_lag_1' not in input_df.columns:
        input_df['responder_6_lag_1'] = 0
    predictions = input_df[['row_id','responder_6_lag_1']].rename({'responder_6_lag_1':'responder_6'},axis=1)

    if isinstance(predictions, pd.DataFrame):
        assert (predictions.columns == ['row_id', 'responder_6']).all()
    else:
        raise TypeError('The predict function must return a DataFrame')
    # Confirm has as many rows as the test data.
    assert len(predictions) == len(test)

    return predictions
```
